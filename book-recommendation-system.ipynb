```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name
```

<div style="background-color: #008000; padding: 20px; border-radius: 10px;">
    <h1 style="text-align: left; font-size: 22px;"> Merges the ratings with book details and filters users who have rated more than 50 books and books rated at least 10 times.</h1>
</div>

```python
# Filter users who have rated more than 50 books
user_ratings_count = ratings_with_name.groupby('User-ID').count()['Book-Rating']
active_users = user_ratings_count[user_ratings_count > 50].index
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
# Filter books that have been rated at least 10 times
book_ratings_count = filtered_ratings.groupby('Book-Title').count()['Book-Rating']
popular_books = book_ratings_count[book_ratings_count >= 10].index
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(popular_books)]
final_ratings
```

<div style="background-color: #ff0000; padding: 20px; border-radius: 10px;">
    <h1 style="text-align: left; font-size: 32px;">User-Item Pivot Table </h1>
</div>

<div style="background-color: #008000; padding: 20px; border-radius: 10px;">
    <h1 style="text-align: left; font-size: 22px;"> Making the table which consist book-title in row,user in column and
ratings are values.</h1>
</div>

```python
pivot_table = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pivot_table.fillna(0, inplace=True)
pivot_table
```

<div style="background-color: #ff0000; padding: 20px; border-radius: 10px;">
    <h1 style="text-align: left; font-size: 32px;">Recommendation using Cosine similarity
</div>

<div style="background-color: #ffA500; padding: 20px; border-radius: 10px;">
    <h1 style="text-align: left; font-size: 32px; color: white;">Cosine Similarity in Collaborative Filtering</h1>
    <p style="color: white; font-size: 18px;">
        Cosine Similarity is a measure of similarity used in collaborative filtering.
        <br>
        It works by measuring the cosine of the angle between two vectors.
        <br>
        In this case, the vectors represent the behavior of two users, and the cosine similarity measures how similar their behavior is.
    </p>
</div>


```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pivot_table)
```

```python
def recommend(book_name):
    index = np.where(pivot_table.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]
    for i in similar_items:
        book_title = pivot_table.index[i[0]]
        print(book_title)

```

```python
recommend('Animal Farm')
```

    <h1 style="text-align: left; font-size: 32px;">KNN(Brute Algorithm and Cosine Similarity)</h1>
</div>

<div style="background-color: #66bbf1; padding: 20px; border-radius: 10px;">
    <h1 style="text-align: left; font-size: 32px; color: white;">Brute-Force Approach in KNN</h1>
    <p style="color: white; font-size: 18px;">
        The brute-force approach in KNN focuses on the training phase, where the distances between data points are calculated and stored.
        <br><br>
        During the prediction phase, these precomputed distances are leveraged to find the nearest neighbors efficiently.
        <br><br>
        <b>Steps:</b>
        <br>
        1. Converts the pivot table into a sparse utility matrix using `csr_matrix` from SciPy.
        <br>
        2. Initializes a `NearestNeighbors` model using the brute-force algorithm (`algorithm='brute'`), comparing all pairs of points for nearest neighbor computation, suitable for smaller datasets.
        <br>
        3. Employs cosine similarity (`metric='cosine'`) to measure the cosine of the angle between vectors, facilitating similarity calculations among different items in the utility matrix.
    </p>
</div>

```python
# Create a utility matrix from final_ratings pivot table
pt_matrix = csr_matrix(pivot_table.values)
print(pt_matrix.toarray())
# Convert the pivot table into a sparse matrix
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(pt_matrix)
```

<div style="background-color: #ff0000; padding: 20px; border-radius: 10px;">
    <h1 style="text-align: left; font-size: 32px;">Function for Recomendation</h1>
</div>

```python
def recommend(book):
    try:
        query_index = pivot_table.index.get_loc(book)
        distances, indices = model_knn.kneighbors(pt_matrix[query_index], n_neighbors=6)

        print(f"Recommendations for '{pivot_table.index[query_index]}':")
        for i in range(1, len(distances.flatten())):
            print(f"{i}: {pivot_table.index[indices.flatten()[i]]}")

    except KeyError:
        print(f"The book '{book}' is not found in the dataset.")
        similar_books = find_similar_books(book)

        if similar_books:
            print("*********************************************")
            print(f"Did you mean one of these books?")
            for i, similar_book in enumerate(similar_books, start=1):
                print(f"{i}: {similar_book}")
        else:
            print(f"No similar books found. Please try a different book name.")



def find_similar_books(book):
    similar_books = [b for b in pivot_table.index if book.lower() in b.lower()]
    return similar_books[:10] if similar_books else []


    
# while True:
#     print("*********************************************")
#     book_name = input("Enter the Book Name (Type 'exit' to stop): ")

#     if book_name.lower() == 'exit':
#         print("Exiting the recommendation system.")
#         print("*********************************************")
#         break

#     recommend(book_name)
```
